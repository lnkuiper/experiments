```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def load_file(fname):
    with open(fname, 'r') as f:
        return np.array([float(l.split(' ')[1].strip())/1000 for l in f.readlines()[2:]])
```

```python
def create_plot(times):
    freq, time = np.histogram(times, bins=20)
    corr_time = []
    for i in range(len(time) - 1):
        corr_time.append((time[i] + time[i+1])/2)
    plt.figure(figsize=(5, 4))
    plt.plot(corr_time, freq)
    plt.xlabel('Retrieval Time (s)')
    plt.ylabel('Number of Queries')
    plt.title('1 Thread')
    plt.savefig('1.png')
```

```python
create_plot(load_file('final_qtime_1'))
```
